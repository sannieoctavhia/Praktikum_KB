```python
cani.set_index('date_review', inplace=True)
cani.head()
```

# Kedua, kita menggunakan variabel date_review sebagai index agar memudahkan kita untuk menggunakan .loc nantinya

```python
job_review = cani['overall_rating'].value_counts()
job_review
```

```python
job_group = cani.groupby('overall_rating').sum()
job_group.head()
```

## Membuat variabel job_review untuk memudahkan pemanggilan saat membuat visualisasi data

```python
# pandas
job_review.plot.bar()
plt.title('Data People Job Review')
plt.xlabel('People Review')
plt.ylabel('Total Review')
plt.show()
```

## 1) Pada grafik batang dapat dilihat bahwa rating masing-masing orang pada 'People Review' (Job) yang pernah mereka jalani sebelumnya maupun yang dijalani sekarang dominasi memberikan rating 4 yang artinya bahwa pekerjaan yang mereka ambil tersebut sangat worth it dengan gaji yang besar dan lingkup kerja yang luas.
## 2) Sedangkan Total Review menampilkan jumlah orang yang telah memberikan reviewnya bagi pekerjaan mereka sendiri

```python
# matplotlib
plt.hist(cani['overall_rating'])
plt.title('Data Job Review')
plt.xlabel('People Review')
plt.ylabel('Total Review')
plt.show()
```

## 1) Pada grafik Histogram ini ditampilkan data 'People Review' pada overall_rating yaitu orang-orang yang memberikan ratingnya bagi pekerjaan yang telah mereka jalani sebelumnya ataupun pekerjaan yang sedang dijalani sekarang ini memberikan rating 4.0 yang dimana artinya banyak pekerjaan yang dapat dikatakan layak untuk dijalani dengan lapangan kerja yang pas, gaji yang lumayan, ruang lingkup pekerjaan yang pas dan pekerjaan yang dirasa cocok dengan kebutuhan masing-masing pereview
## 2) Dengan Total review mencapai 250000 orang yang memberikan rating pada pekerjaan mereka dengan rating 4.0

```python
plt.figure(figsize=(7,12))
job_review['work_life_balance'].plot.pie(autopct='%.1f%%')
plt.title('Total Review jika dipersentasekan')
plt.show()
```

## terlihat dari diagram pie bahwa kebanyakan orang memberikan rating yang cukup baik bagi pekerjaannya 

```python
# seaborn
sns.scatterplot(
    data=cani,
    x='job_title',
    y='work_life_balance',
    hue='overall_rating'
)

plt.title('Data Job Review Menurut Pekerjaan')
plt.xlabel('Job')
plt.ylabel('Worth for Future')
plt.show()
```

## Pada grafik Scatterplot, menunjukkan bahwa banyak jenis pekerjaan yang sesuai dengan apa yang mereka inginkan dengan gaji yang cukup, ruang lingkup pekerjaan yang cukup memuaskan

```python
job_review.loc[:, 'overall_rating':'recommend'].plot()
plt.title('Data Job Review menurut recommend dari orang-orang menggunakan grafik lainnya')
plt.xlabel('JOB REVIEW')
plt.ylabel('Jumlah Recommend')
plt.show()
```

## Dapat dilihat dari grafik diatas bahwa banyak dari per 70% orang memberikan recommended bagi pekerjaan yang mereka jalani sekarang 

```python
# seaborn
total_corr = cani.loc[:,'date_review':'recommend'].corr()
total_corr
```

```python
sns.heatmap(total_corr, annot=True, cmap='cividis')
```
